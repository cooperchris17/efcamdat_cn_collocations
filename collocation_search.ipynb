# This is a simple notebook to search for sentences 
### based on the verb or the verb and noun in the collocations

```python
import pandas as pd
import numpy as np

df = pd.read_csv('verb_noun_collocations_27May.csv')
df
```

# Sentence search for only 'verb'

```python
verb = 'sale'

# Filtering sentences where 'verb' column is 'ale'
filtered_sentences = df[df['verb'] == verb]['sentence']

# Randomly selecting 20 sentences
random_sentences = np.random.choice(filtered_sentences, size=20, replace=False)

# Printing the sentences
for sentence in random_sentences:
    print(sentence)
```

# Sentence search for 'verb' and 'noun'

```python
# Define search criteria
verb = 'have'
noun = 'hair'

# Filtering sentences where 'verb' column is 'sale' AND 'noun' column is 'sale'
filtered_sentences = df[(df['verb'] == verb) & (df['noun'] == noun)]['sentence']

# Randomly selecting 20 sentences
random_sentences = np.random.choice(filtered_sentences, size=20, replace=False)

# Printing the sentences
for sentence in random_sentences:
    print(sentence)

```
