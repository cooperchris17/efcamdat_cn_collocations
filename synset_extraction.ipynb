# Extract polysemy information for the verbs
### This is what the synset info looks like in Wordnet

```python
from nltk.corpus import wordnet as wn
import pandas as pd

# for POS (nouns)
wn.synsets('book', pos=wn.NOUN)
```

```python
# for POS (verbs)
wn.synsets('book', pos=wn.VERB)
```

```python
# no POS
wn.synsets('book')
```

# Now run the code on the verb list

```python
verb_count_df = pd.read_csv('verb_counts_27May.csv') 
verb_count_df
```

```python
verb_count_df['verb']
```

```python
# if any values in the df were 'NaN', remove (it will cause an error)
df = verb_count_df.dropna(subset=['verb'])
df
```

```python
df = df.copy()

# calculate the number of verb synsets for each verb
df['v_syn_count'] = df['verb'].apply(lambda x: len(wn.synsets(x, pos=wn.VERB)))
# store the actual verb synsets for each verb
df['v_synsets'] = df['verb'].apply(lambda x: [synset.name() for synset in wn.synsets(x, pos=wn.VERB)])
# calculate the number of all synsets for each verb
df['syn_count'] = df['verb'].apply(lambda x: len(wn.synsets(x)))
# store the all synsets for each verb
df['synsets'] = df['verb'].apply(lambda x: [synset.name() for synset in wn.synsets(x)])

df.to_csv('synsets_27May.csv', index=False)
df
```

```python
# show all the rows where v_syn_count is 0
zero_v_synsets = df[df['v_syn_count'] == 0]
zero_v_synsets.to_csv('zero_v_synsets_27May.csv', index=False)
zero_v_synsets
```
