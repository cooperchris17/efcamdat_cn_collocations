## 1. Standardize spelling to American English 
## 2. Minimal spelling correction to account for possible typos (as described in the paper)

```python
import pandas as pd

df = pd.read_csv('verb_noun_collocations_22May.csv')
efcam = df.dropna().copy()
efcam
```

```python
from breame.spelling import british_spelling_exists, get_american_spelling

def standardize_word_to_american(word):
    """
    Convert a British English word to American English.
    Checks if the word is British English, then converts if it is.
    """
    if word is None or pd.isna(word):
        return word
        
    # Check if this word is recognized as British English
    if british_spelling_exists(word):
        # Convert to American spelling
        return get_american_spelling(word)
    else:
        # Keep the original word if it's not recognized as British
        return word

# Apply function to the 'verb' and 'noun' columns in EFCAMDAT
efcam.loc[:, 'verb'] = efcam['verb'].apply(standardize_word_to_american)
efcam.loc[:, 'noun'] = efcam['noun'].apply(standardize_word_to_american)
efcam
```

# Prepare proper nouns from Paul Nation's list
### this will reduce the chance of words being corrected as proper nouns (previous attempt included corrections such as adam and belfast)
#### Nation, I.S.P. (2017). The BNC/COCA Level 6 word family lists (Version 1.0.0) [Data file]. Available from http://www.victoria.ac.nz/lals/staff/paul-nation.aspx

```python
# read the txt file basewrd31_proper_nouns.txt and print the contents
# remove tabs and lowercase the contents
# add each line to an item # in a list called proper_nouns
file_path = 'basewrd31_proper_nouns.txt'

with open(file_path, 'r') as file:
    contents = file.read()
    contents = contents.replace('\t', '').lower()
    proper_nouns = contents.splitlines()
    proper_nouns_set = set(proper_nouns)  # Convert to a set to potentially speed up lookups

# print the first 100 proper nouns
proper_nouns_set
```

## Using pyspellcheck to identify misspellings and suggest candidates (Edit Distance / Levensthein Distance <=2)
### Use Wordnet to only offer corrections in same POS
### Also choose candidates based on frequency (New-GSL (Gablasova & Brezina, 2015), then wordfreq package)
### Also don't correct to proper noun
### Some manual corrections & words not to be corrected are dealt with at the start of this block

```python
import re
import string
import pandas as pd
from collections import defaultdict
from spellchecker import SpellChecker
from nltk.corpus import wordnet as wn
from wordfreq import zipf_frequency

# -----------------------------------------------
# Manual validation and correction
# -----------------------------------------------

# Checked all 'mispelled verbs' with frequency >= 2
# 'e-mail' was the only one that I could find
valid_verbs = ['e-mail']

# Manually correct nouns in df['noun']
# 'tvs' and 'cds' were clearly not lemmatized properly (maybe due to lowercasing?)
manual_noun_corrections = {
    'tv.': 'tv',
    'tvs': 'tv',
    'cd.': 'cd',
    'cds': 'cd',
}

# Manual corrections for verbs
manual_verb_corrections = {
    'paly': 'play'
}

# Checked all misspelled nouns with frequency >= 2
# 'rmb' is currency, always preceded by a number (maybe not in BNC, but kept anyway)
# Some of these words are likely not in the dictionary because they are recent
# They will also not be in the BNC, but we might flag them manually as acceptable
# 'etc.' et cetera is classified as a noun in some dictionaries, e.g. Oxford Learner's Dictionary
valid_nouns = [
    'tv', 'e-mail', 'mp3', 'etc.', 'non-stop', 'e-ticket', 'smartphone', 'unsatisfaction',
    'cv', 'rmb', 'mba', 'pc', 'ceo', 'fiance', 'dvd', 'mini-skirt', 'bbq', 'e-book',
    'wifi', 'app', 'toefl', 'ef', 'mooncake', 'a.m.', 'p.m.', 'cbt', 'mangoe', 'backsplash',
    'etc.', 'cctv', 'diy', 
]

# -----------------------------------------------
# Double-check lowercasing and stripping
# -----------------------------------------------

# Normalize case and strip whitespace
efcam['verb'] = efcam['verb'].str.lower().str.strip()
efcam['noun'] = efcam['noun'].str.lower().str.strip()

# -----------------------------------------------
# Build Wordnet vocabulary sets
# -----------------------------------------------

# Extract all lemma names for nouns and verbs
wordnet_nouns = {lemma.name().lower() for syn in wn.all_synsets('n') for lemma in syn.lemmas()}
wordnet_verbs = {lemma.name().lower() for syn in wn.all_synsets('v') for lemma in syn.lemmas()}

# Combine with manually validated words
noun_vocab = wordnet_nouns.union(valid_nouns)
verb_vocab = wordnet_verbs.union(valid_verbs)


spell_noun = SpellChecker(distance=2)
spell_noun.word_frequency.load_words(noun_vocab)

spell_verb = SpellChecker(distance=2)
spell_verb.word_frequency.load_words(verb_vocab)

# -----------------------------------------------
# Load NGSL lists for high frequency candidates
# -----------------------------------------------

ngsl_df = pd.read_csv("NewGSL_ame.csv")

# Lowercase headwords just in case
ngsl_df['Headword'] = ngsl_df['Headword'].str.lower()

# Filter by part of speech
ngsl_nouns = ngsl_df[ngsl_df['pos'] == 'n']
ngsl_verbs = ngsl_df[ngsl_df['pos'] == 'v']

# Build rank dictionaries
ngsl_noun_rank = dict(zip(ngsl_nouns['Headword'], ngsl_nouns['Rank']))
ngsl_verb_rank = dict(zip(ngsl_verbs['Headword'], ngsl_verbs['Rank']))


# -----------------------------------------------
# Spell-correct a single token with overrides
# -----------------------------------------------

def sanitize_word(word):
    # Remove leading/trailing punctuation and stray non-word characters
    word = word.strip().lower()
    word = re.sub(r"^[^a-zA-Z]+", "", word)  # leading non-letters
    word = re.sub(r"[^a-zA-Z]+$", "", word)  # trailing non-letters
    word = word.replace("'", "")  # remove inner apostrophes
    return word

def correct_token(word, wordnet_vocab, valid_list, correction_dict, spellchecker, pos_tag, proper_nouns):
    if not isinstance(word, str):
        return word

    raw_word = word
    word = sanitize_word(word)
    if word == "":
        return raw_word

    word = correction_dict.get(word, word)

    # Skip correction if it's a proper noun
    if word in proper_nouns:
        return raw_word

    if word in valid_list or (word in wordnet_vocab and word not in proper_nouns):
        return word

    candidates = spellchecker.candidates(word)
    if candidates:
        valid_cands = [cand for cand in candidates if cand in wordnet_vocab and cand not in proper_nouns]
        if valid_cands:
            # use NGSL highest rank words first, if not in NGSL, use Zipf frequency
            # Define rank source
            ngsl_rank = ngsl_noun_rank if pos_tag == 'n' else ngsl_verb_rank

            def score(w):
                return (
                    -ngsl_rank.get(w, float('inf')),  # NGSL rank: lower is better
                    zipf_frequency(w, "en", wordlist="best")  # Zipf frequency as tie-breaker
                )

            return max(valid_cands, key=score)

    return raw_word


# -----------------------------------------------
# Correct verb-noun collocation pair
# -----------------------------------------------

def correct_collocation_spellchecker(verb, noun):
    original_verb = verb
    original_noun = noun

    corrected_verb = correct_token(
    verb,
    wordnet_verbs,
    valid_verbs,
    manual_verb_corrections,
    spell_verb,
    'v',
    proper_nouns_set
)

    corrected_noun = correct_token(
    noun,
    wordnet_nouns,
    valid_nouns,
    manual_noun_corrections,
    spell_noun,
    'n',
    proper_nouns_set
)

    was_corrected = (corrected_verb != verb or corrected_noun != noun)
    return corrected_verb, (verb if corrected_verb != verb else "-"), corrected_noun, (noun if corrected_noun != noun else "-"), was_corrected

# -----------------------------------------------
# Apply corrections to learner corpus
# -----------------------------------------------

corrections = efcam.apply(
    lambda row: correct_collocation_spellchecker(row['verb'], row['noun']),
    axis=1,
    result_type='expand'
)

# Rename for clarity
corrections.columns = ['corrected_verb', 'original_verb', 'corrected_noun', 'original_noun', '_was_corrected']

# Replace original verb/noun with corrected versions
efcam['verb'] = corrections['corrected_verb']
efcam['noun'] = corrections['corrected_noun']

# Insert original_verb after 'verb'
verb_index = efcam.columns.get_loc('verb')
efcam.insert(verb_index + 1, 'original_verb', corrections['original_verb'])

# Insert original_noun after 'noun'
noun_index = efcam.columns.get_loc('noun')
efcam.insert(noun_index + 1, 'original_noun', corrections['original_noun'])
efcam
```
